```python
import threading
import time
import random
rgb_modules = list(set([".".join(ch.split('.')[0:3]) for ch in ch_names if 'ESH10000355.A1' in ch]))
colors = ['White', 'Black', 'Red', 'Green', 'Blue', 'Pink']
try:
    stop.release()
except:
    stop = threading.Event()
def blink():
    stop.release()
    while not stop.is_set():
        for rgb_module in rgb_modules:
            for connector in 'ABCDEF':
                color = random.choice(colors)
                # print(rgb_module, connector, color)
                acc.set_value(rgb_module+'.'+connector+'1', color)
                #time.sleep(0.15)
```

```python
threading.Thread(target=blink, daemon=True).start()
```
